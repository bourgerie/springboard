```python
df_train = df_train.drop('keyword', axis=1)
df_train.drop('text', axis=1, inplace=True)
df_train.drop('text_clean', axis=1, inplace=True)
```

```python
df_train['target'] = df_train['target'].replace(1, "True")
df_train['target'] = df_train['target'].replace(0, "False")
```

```python
df_train.head()
```

# Split test and train sets


```python
y = df_train['target'] #label of whether it's a disaster or not
X = df_train.drop('target', axis=1) #tweet data, already been vectorized
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(X, y, test_size = 0.25)
```

# Bayes

```python
from sklearn.naive_bayes import MultinomialNB

naive_bayes_classifier = MultinomialNB()
naive_bayes_classifier.fit(X_train, y_train)

#https://iq.opengenus.org/naive-bayes-on-tf-idf-vectorized-matrix/
```

```python
from sklearn import metrics

y_pred = naive_bayes_classifier.predict(X_test)

# compute the performance measures
score = metrics.accuracy_score(y_test, y_pred)
print("accuracy:   %0.25f" % score)
#is that the right? string?

print(metrics.classification_report(y_test, y_pred,target_names=['True', 'False']))


print(metrics.confusion_matrix(y_test, y_pred))
```

# SVM

```python
#Import svm model
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') 

clf.fit(X_train, y_train)
y_pred2 = clf.predict(X_test)
```

```python
#metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred2))
```

```python
print(metrics.classification_report(y_test, y_pred2,target_names=['True', 'False']))


print(metrics.confusion_matrix(y_test, y_pred2))
```
